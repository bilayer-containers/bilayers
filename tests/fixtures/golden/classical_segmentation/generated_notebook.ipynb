````
from ipywidgets import widgets, VBox, HBox, Tab, Accordion
from IPython.display import display
from typing import Any, Optional, Union
import os, tempfile


inputs = {'input_images': {'name': 'input_images', 'type': 'image', 'label': 'Drag and drop all images to be analyzed', 'subtype': ['grayscale', 'color', 'binary'], 'description': 'Upload all images to be analyzed', 'cli_tag': '--folder', 'cli_order': 0, 'default': 'directory', 'optional': False, 'format': ['tiff', 'ometiff', 'png', 'jpg', 'jpeg'], 'folder_name': '/bilayers/input_images', 'file_count': 'multiple', 'unique_string': ['*'], 'section_id': 'inputs', 'mode': 'beginner', 'depth': True, 'timepoints': True, 'tiled': True, 'pyramidal': True}}
outputs = {'segmented_images': {'name': 'segmented_images', 'type': 'image', 'label': 'Segmented Images', 'subtype': ['label'], 'description': 'Segmented images', 'cli_tag': 'None', 'cli_order': 0, 'default': 'directory', 'optional': False, 'format': ['tiff'], 'folder_name': '/bilayers/output_images', 'file_count': 'single', 'unique_string': ['_output'], 'section_id': 'outputs', 'mode': 'beginner', 'depth': True, 'timepoints': True, 'tiled': True, 'pyramidal': True}}
parameters = {'threshold_method': {'name': 'threshold_method', 'type': 'radio', 'label': 'Select a Threshold Method', 'description': 'Select a threshold method to segment the image', 'options': [{'label': 'otsu', 'value': 'otsu'}, {'label': 'li', 'value': 'li'}], 'default': 'otsu', 'cli_tag': '--threshold_method', 'optional': False, 'section_id': 'input-args', 'mode': 'beginner'}, 'min_diameter': {'name': 'min_diameter', 'type': 'float', 'label': 'Object Minimum Diameter Size', 'description': 'Minimum diameter of objects in pixels', 'default': 5, 'cli_tag': '--min_size', 'optional': False, 'section_id': 'input-args', 'mode': 'beginner'}, 'max_diameter': {'name': 'max_diameter', 'type': 'float', 'label': 'Object Maximum Diameter Size', 'description': 'Maximum diameter of objects in pixels', 'default': 20, 'cli_tag': '--max_size', 'optional': False, 'section_id': 'input-args', 'mode': 'beginner'}, 'save_dir': {'name': 'save_dir', 'type': 'textbox', 'label': 'Save Directory', 'description': 'directory to save output files', 'output_dir_set': True, 'default': '/bilayers/output_images', 'cli_tag': '--save_dir', 'optional': True, 'section_id': 'output-args', 'mode': 'advanced'}}
display_only = {}
exec_function = {'name': 'generate_cli_command', 'cli_command': 'python -m classical_segmentation', 'hidden_args': {}}
cli_sequence = [{"cli_order": 0, "cli_tag": "--folder", "default": "directory", "depth": True, "description": "Upload all images to be analyzed", "file_count": "multiple", "folder_name": "/bilayers/input_images", "format": ["tiff", "ometiff", "png", "jpg", "jpeg"], "label": "Drag and drop all images to be analyzed", "mode": "beginner", "name": "input_images", "optional": False, "pyramidal": True, "section_id": "inputs", "source": "input", "subtype": ["grayscale", "color", "binary"], "tiled": True, "timepoints": True, "type": "image", "unique_string": ["*"]}, {"cli_tag": "--threshold_method", "default": "otsu", "description": "Select a threshold method to segment the image", "label": "Select a Threshold Method", "mode": "beginner", "name": "threshold_method", "optional": False, "options": [{"label": "otsu", "value": "otsu"}, {"label": "li", "value": "li"}], "section_id": "input-args", "source": "parameter", "type": "radio"}, {"cli_tag": "--min_size", "default": 5, "description": "Minimum diameter of objects in pixels", "label": "Object Minimum Diameter Size", "mode": "beginner", "name": "min_diameter", "optional": False, "section_id": "input-args", "source": "parameter", "type": "float"}, {"cli_tag": "--max_size", "default": 20, "description": "Maximum diameter of objects in pixels", "label": "Object Maximum Diameter Size", "mode": "beginner", "name": "max_diameter", "optional": False, "section_id": "input-args", "source": "parameter", "type": "float"}, {"cli_tag": "--save_dir", "default": "/bilayers/output_images", "description": "directory to save output files", "label": "Save Directory", "mode": "advanced", "name": "save_dir", "optional": True, "output_dir_set": True, "section_id": "output-args", "source": "parameter", "type": "textbox"}]



beginner_widgets = []
advanced_widgets = []

param_vars = {}

# Appending args in cli_command
cli_command_base = exec_function.get('cli_command', "")

radio_options = {

    "threshold_method": {
            "otsu": "otsu", "li": "li"
        },
    
    }







mount_path = "/bilayers/input_images"
upload_temp_dir = tempfile.mkdtemp(prefix="jupyter_upload_")

# Check if mount folder exists and has files
if os.path.exists(mount_path) and any(os.scandir(mount_path)):
    # If mounted, display info and disable upload
    input_images = widgets.HTML(
        value=(
            f"<div style='padding:8px; background:#eaffea; border-radius:6px;'>"
            f"<b>Using Mounted Directory:</b> <code>{mount_path}</code><br>"
            f"<i>File upload is disabled for this input.</i>"
            f"</div>"
        )
    )
    current_input_dir = mount_path
    is_mounted = True
else:
    # Otherwise, enable the FileUpload widget
    input_images = widgets.FileUpload(multiple=True, description="Upload Files")
    current_input_dir = upload_temp_dir
    is_mounted = False

# Add a description label
desc_input_images  = widgets.HTML(value = "<b>Drag and drop all images to be analyzed </b> :  Upload all images to be analyzed")

# Add widget and description to accordion
tab = Tab(children=[HBox([widgets.Label("Drag and drop all images to be analyzed"), input_images]), desc_input_images])
tab.set_title(0, "Drag and drop all images to be analyzed")
tab.set_title(1, "Description")

# Append the corresponding cli_tag, widget to param_vars

param_vars["input_images"] = {
    "cli_tag": "--folder",
    "widget": input_images if not is_mounted else widgets.Label(value=f"Mounted: {mount_path}"),
    "cli_order": 0,
    "folder_name": current_input_dir,
    "optional": False,
    "is_mounted": is_mounted
}


beginner_widgets.append(tab)








threshold_method = widgets.RadioButtons(options=[("otsu", "otsu"), ("li", "li")], value="otsu")


# Add a description label
desc_threshold_method  = widgets.HTML(value = "<b>Select a Threshold Method </b> :  Select a threshold method to segment the image")

# Add widget and description to accordion
tab = Tab(children=[HBox([widgets.Label("Select a Threshold Method"), threshold_method]), desc_threshold_method])
tab.set_title(0, "Select a Threshold Method")
tab.set_title(1, "Description")

# Append the corresponding cli_tag, widget to param_vars

param_vars["threshold_method"] = {
    "cli_tag": "--threshold_method",
    "widget": threshold_method,
    "cli_order": 0,
    "optional": False
}



beginner_widgets.append(tab)



 
min_diameter = widgets.FloatText(value=5)



# Add a description label
desc_min_diameter  = widgets.HTML(value = "<b>Object Minimum Diameter Size </b> :  Minimum diameter of objects in pixels")

# Add widget and description to accordion
tab = Tab(children=[HBox([widgets.Label("Object Minimum Diameter Size"), min_diameter]), desc_min_diameter])
tab.set_title(0, "Object Minimum Diameter Size")
tab.set_title(1, "Description")

# Append the corresponding cli_tag, widget to param_vars

param_vars["min_diameter"] = {
    "cli_tag": "--min_size",
    "widget": min_diameter,
    "cli_order": 0,
    "optional": False
}



beginner_widgets.append(tab)



 
max_diameter = widgets.FloatText(value=20)



# Add a description label
desc_max_diameter  = widgets.HTML(value = "<b>Object Maximum Diameter Size </b> :  Maximum diameter of objects in pixels")

# Add widget and description to accordion
tab = Tab(children=[HBox([widgets.Label("Object Maximum Diameter Size"), max_diameter]), desc_max_diameter])
tab.set_title(0, "Object Maximum Diameter Size")
tab.set_title(1, "Description")

# Append the corresponding cli_tag, widget to param_vars

param_vars["max_diameter"] = {
    "cli_tag": "--max_size",
    "widget": max_diameter,
    "cli_order": 0,
    "optional": False
}



beginner_widgets.append(tab)



save_dir = widgets.Textarea(value="/bilayers/output_images")


# Add a description label
desc_save_dir  = widgets.HTML(value = "<b>Save Directory </b> :  directory to save output files")

# Add widget and description to accordion
tab = Tab(children=[HBox([widgets.Label("Save Directory"), save_dir]), desc_save_dir])
tab.set_title(0, "Save Directory")
tab.set_title(1, "Description")

# Append the corresponding cli_tag, widget to param_vars

param_vars["save_dir"] = {
    "cli_tag": "--save_dir",
    "widget": save_dir,
    "cli_order": 0,
    "optional": True
}




advanced_widgets.append(tab)






acc = Accordion()
# Add beginner_widgets to the first section
acc.children = [VBox(beginner_widgets)]
acc.set_title(0, "Beginner's Section")
# Add advanced_widgets to the second section
acc.children = acc.children + (VBox(advanced_widgets),)
acc.set_title(1, "Advanced Section")

def option_to_append(cli_tag: str, value: Any) -> Optional[str]:
    """
    Formats CLI options for appending to the command.

    Args:
        cli_tag (str): CLI tag (e.g., "--option").
        value (Any): The value to append.

    Returns:
        Optional[str]: The formatted CLI option or None if ignored.
    """
    if cli_tag == "None":
        return None # Ignore the cli_tag and value pair
    elif cli_tag == "":
        return str(value)  # Append only the value
    else:
        if "=" in cli_tag:
            return f"{cli_tag}{value}"  # Append cli_tag and value with "="
        return f"{cli_tag} {value}"  # Append cli_tag and value

def validate_and_highlight(widget: widgets.Widget, value: Any, optional: bool) -> bool:
    """
    Validates whether the given widget's value is present for non-optional fields.
    If the value is missing (None, empty, or the string "None") for a mandatory field,
    it sets the widget's border to red to visually indicate an error.
    
    Args:
        widget (widgets.Widget): The widget to validate.
        value (Any): The value of the widget.
        optional (bool): Whether the field is optional.
    
    Returns:
        bool: True if the value is valid (or the field is optional), False otherwise.
    """
    if (value is None or value == "" or value == "None") and not optional:
        widget.layout.border = '2px solid red'
        return False
    else:
        widget.layout.border = ''
        return True

def handle_file_upload(change: Any) -> None:
    """
    Handles file upload events by saving uploaded files to disk
    and refreshing the CLI command.

    This ensures FileUpload widget data (stored in memory) is
    persisted to the temp or mounted folder before running CLI.

    See also:
    Resolves: https://github.com/bilayer-containers/bilayers/issues/118
    """
    widget = change["owner"]

    # Find which param this widget belongs to
    for key, param_info in param_vars.items():
        if param_info.get("widget") is widget:
            folder_name = param_info.get("folder_name", tempfile.mkdtemp(prefix="bilayers_upload_"))
            os.makedirs(folder_name, exist_ok=True)

            # ipywidgets v8: .value is a list of dicts
            for file_data in widget.value:
                file_name = file_data.get("name")
                file_content = file_data.get("content")
                if file_name and file_content:
                    save_path = os.path.join(folder_name, file_name)
                    with open(save_path, "wb") as f:
                        f.write(file_content)
                    print(f"Saved uploaded file: {save_path}!")

            # Update CLI after saving files
            cli_command.value = construct_cli_command()
            break

def construct_cli_command() -> str:
    """
    Constructs the CLI command dynamically based on selected widgets.

    Returns:
        str: The constructed CLI command string.
    """
    cli_command_parts = [cli_command_base]

    for item in cli_sequence:
        name = item.get("name", "")
        source = item.get("source", "")
        cli_tag = item.get("cli_tag", "")
        optional = item.get("optional", False)
        append_value = item.get("append_value", False)

        if source == "hidden":
            value = item.get("value", "")
            if isinstance(value, bool):
                if append_value:
                    option = option_to_append(cli_tag, value)
                    if option is not None:
                        cli_command_parts.append(option)
                else:
                    if value and cli_tag != "None":
                        cli_command_parts.append(cli_tag)
            else:
                option = option_to_append(cli_tag, value)
                if option is not None:
                    cli_command_parts.append(option)
            continue

        param_key = name.lower().replace(" ", "_")
        param_info = param_vars.get(param_key)
        if not param_info:
            continue

        widget: Union[widgets.Widget, widgets.FileUpload, widgets.Checkbox] = param_info["widget"]
        optional = param_info.get("optional", False)

        # =====================================================
        # Case A: Folder-based input (image, file, etc.)
        # =====================================================
        if "folder_name" in param_info:
            folder_name = param_info["folder_name"]
            is_mounted = param_info.get("is_mounted", False)
            widget = param_info.get("widget")

            if is_mounted:
                value = folder_name
            elif isinstance(widget, widgets.FileUpload) and widget.value:
                os.makedirs(folder_name, exist_ok=True)
                for file_info in widget.value:
                    fname = file_info.get("name")
                    content = file_info.get("content")
                    if fname and content:
                        with open(os.path.join(folder_name, fname), "wb") as f:
                            f.write(content)
                print(f"Uploaded files saved to: {folder_name}")
                value = folder_name
            else:
                print(f"No input files found for {cli_tag}")
                continue

            option = option_to_append(cli_tag, value)
            if option is not None:
                cli_command_parts.append(option)
            continue

        # =====================================================
        # Case B: Checkbox
        # =====================================================
        if isinstance(widget, widgets.Checkbox):
            append_value = param_info.get("append_value", False)
            value = widget.value
            if append_value:
                option = option_to_append(cli_tag, value)
                if option is not None:
                    cli_command_parts.append(option)
            else:
                if value and cli_tag != "None":
                    cli_command_parts.append(cli_tag)
            continue

        # =====================================================
        # Case C: Other widgets
        # =====================================================
        if isinstance(widget, widgets.RadioButtons):
            value = widget.value
            if value == "ignore":
                value = None
        else:
            value = widget.value

        validate_and_highlight(widget, value, optional)
        if value in [None, "", "None"]:
            continue

        option = option_to_append(cli_tag, value)
        if option is not None:
            cli_command_parts.append(option)

    return " ".join(cli_command_parts)

def update_command(change: Any) -> None:
    """
    Updates the CLI command when a widget value changes.

    Args:
        change (Any): The event triggering the update.
    """
    cli_command.value = construct_cli_command()

cli_command: widgets.Textarea = widgets.Textarea(value=construct_cli_command(), description="CLI Command", layout=widgets.Layout(width="90%"))

for key, param_info in param_vars.items():
    widget = param_info["widget"]

    # Case A: FileUpload widget --> use custom handler
    if isinstance(widget, widgets.FileUpload):
        widget.observe(handle_file_upload, names='value')

    # Case B: all other widgets --> use normal observer
    elif hasattr(widget, 'observe'):
        widget.observe(update_command, names='value')

if __name__ == "__main__":
    from IPython.display import display
    display(acc, cli_command)
````

````
def final_validation() -> None:
    """
    Checks that all mandatory parameters have valid (non-empty) values.
    Raises an Exception if any non-optional parameter is missing.
    """
    for key, param_info in param_vars.items():
        cli_tag: str = param_info["cli_tag"]
        widget: Union[widgets.Widget, widgets.FileUpload, widgets.Checkbox] = param_info["widget"]
        optional: bool = param_info["optional"]
        value = widget.value
        
        # For RadioButtons, convert "ignore" to None
        if isinstance(widget, widgets.RadioButtons) and value == "ignore":
            value = None
        
        # If the value is missing (None, empty string, or "None") and the parameter is mandatory, raise error
        if ((value is None) or (value == "") or (value == "None")) and (optional is False):
            raise Exception(f"Error: {cli_tag} is mandatory but has no value!")
    
    # If we reach here, all mandatory parameters have values
    print("✔️ Final validation passed: All required fields are set!")

# Run final validation in a separate cell
try:
    final_validation()
except Exception as e:
    print(e)
````
